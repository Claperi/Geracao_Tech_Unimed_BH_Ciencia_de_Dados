```python
# Identificação dos Continentes no dataFrame
df["Continente"].unique()
```

```python
# Retorno de valores a um determinado Continente
# Atribui-se a uma variável chamada Oceania
Oceania = df.loc[df["Continente"] == "Oceania"]
Oceania.head()
```

```python
# Filtrando quais dados compõe a solicitação acima
Oceania["Continente"].unique() 
```

```python
# Agrupamento de dados
# Quantos paises há em cada Continente
df.groupby("Continente")["Pais"].nunique()
```

```python
# Para cada Ano qual a Expectativa de vida média no Conjunto de Dados
df.groupby("Ano")["Expectativa de Vida"].mean()
```

```python
# Calculando a média do PIB
df["PIB"].mean()
```

```python
# Somando todos os PIB
df["PIB"].sum()
```
